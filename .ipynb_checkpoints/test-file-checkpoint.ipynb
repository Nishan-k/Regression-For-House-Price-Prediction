# Categorical and Numerical Columns

```python
categorical = [col for col in dataset.columns if dataset[col].dtype=='O']
```

```python
categorical
```

```python
numerical = [col for col in dataset.columns if dataset[col].dtypes !='O']
```

```python
numerical
```

**categorical missing values**

```python
def categorical_count(data):
    for col in categorical:
        if (data[col].isnull().sum()>0):
            print('The column {a} has {b} missing values.'.format(a=col, b=data[col].isnull().sum()))
```

```python
categorical_count(dataset)
```

# numerical count

```python
def numerical_count(data):
    for col in numerical:
        if(data[col].isnull().sum()>0):
            print ('The column {a} has missing values of {b}'.format(a=col, b=data[col].isnull().sum()))
```

```python
numerical_count(dataset)
```
